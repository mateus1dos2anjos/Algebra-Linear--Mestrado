```python
import math
```

```python
# Aplica a Eliminação de Gaus para obter o sistema equivalente triagular superior
def eliminationGaussPivPartial(A,b):
    n = len(A)
    
    # Acessar linhas
    for i in range(n):         
        # Verificar qual o maior pivô
        pivo = math.fabs(A[i][i])
        linhaPivo = i
        for j in range(i+1, len(A)):
            if math.fabs(A[j][i]) > pivo:
                pivo = math.fabs(A[j][i])
                linhaPivo = j
        
        # Trocar as linhas 
        if linhaPivo != i:
            linhaAux = A[i]
            A[i] = A[linhaPivo]
            A[linhaPivo] = linhaAux
            
            bAux = b[i]
            b[i] = b[linhaPivo]
            b[linhaPivo] = bAux
            
        # Eliminação de Gauss
        for m in range(i+1, n):
            mult = A[m][i]/A[i][i]
            # Atualizar valores da linha m da matriz A
            for k in range(i, n):
                A[m][k] -= mult*A[i][k]
            # Atualizar valores da linha m de b
            b[m] -= mult*b[i]
    
    return A,b
```

```python
# Resolve sistema triangular superior
def solveUpperTriangularMatrix1(U,b):
    n = len(U)
    x = [0]*n
    x[n-1] = b[n-1]/U[n-1][n-1]
    
    for i in range(n-1,0, -1): # i de n-1 até 1, decrementando 1
        s = 0
        for j in range(i+1,n+1): # j de i+1 até n
            s += U[i-1][j-1]*x[j-1]

        x[i-1] = (b[i-1]-s)/(U[i-1][i-1])

    return x
```

```python
# Aplica a Eliminação de Gaus para obter o sistema equivalente triagular superior
def eliminationGaussPivTotal(A,b):
    n = len(A)
    
    auxPosicoes = []
    for i in range(n):    
        auxPosicoes.append(i)
    #print(auxPosicoes)
    
    # Acessar linhas
    for i in range(n):       
        # Verificar qual o maior pivô
        pivo = math.fabs(A[i][i])
        linhaPivo = i
        for l in range(i, n):
            for j in range(i, n):
                if math.fabs(A[j][l]) != 0:
                    pivo = math.fabs(A[j][i])
                    linhaPivo = j
                    colunaPivo = l
            
        # Trocar as linhas 
        if linhaPivo != i:
            linhaAux = A[i]
            A[i] = A[linhaPivo]
            A[linhaPivo] = linhaAux
            
            bAux = b[i]
            b[i] = b[linhaPivo]
            b[linhaPivo] = bAux
            
        # Trocar as colunas 
        if colunaPivo != i:                   
            colunaAux = []
            for k in range(n):    
                colunaAux.append(A[k][i])
            
            for k in range(n):
                A[k][i] = A[k][colunaPivo]
                A[k][colunaPivo] = colunaAux[k]
            
            # Atualizando índices das incognitas     
            auxPosicoes[i], auxPosicoes[colunaPivo] = auxPosicoes[colunaPivo], auxPosicoes[i]
            # print(auxPosicoes)
                              
        # Eliminação de Gauss
        for m in range(i+1, n):
            mult = A[m][i]/A[i][i]
            # Atualizar valores da linha m da matriz A
            for k in range(i, n):
                A[m][k] -= mult*A[i][k]
            # Atualizar valores da linha m de b
            b[m] -= mult*b[i]
    
    return A,b,auxPosicoes
```

```python
# Resolve sistema triangular superior
def solveUpperTriangularMatrix2(U,b,auxPosicoes):
    n = len(U)
    x = [0]*n
    x[n-1] = b[n-1]/U[n-1][n-1]
    
    for i in range(n-1,0, -1): # i de n-1 até 1, decrementando 1
        s = 0
        for j in range(i+1,n+1): # i de i+1 até n
            s += U[i-1][j-1]*x[j-1]

        x[i-1] = (b[i-1]-s)/(U[i-1][i-1])

    #return x
    for i in range(n):    
        pos = auxPosicoes.index(i)
        print(f'X{auxPosicoes[pos]+1} = {x[pos]}')
```
